```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import datetime
from scipy.stats import norm
import random
```

```python
##Creating Panel Names##
##Twenty Panels in this process##

start_string = "PANEL_"
panels = [ start_string + str(i)  for i in range(50, 70)]
```

```python
##Next, time to define the two machine processes
##We will call each machine as a function, as each machine
##can have a variation of probabilities.

def machine_a():
    time_spent = (60 * 8)
    return time_spent

def machine_b():
    time_spent = (60 * 15)
    return time_spent
```

```python
##Next, we have to start the process of figuring out
##how long each panel spends in machine a.
##Because machine b is dependent on machine a
##Finishing its process before continuing, we
##need to calculate those start times seperately

next_start_time = (60 * 60 * 8) #8am Sharp
process_load_time = 15 #15 seconds to load a machine onces a panel has been ejected

machine_a_start = []
machine_a_end = []

for panel in panels:
    machine_load = next_start_time + process_load_time
    machine_a_start.append(machine_load)
    machine_a_process_time = machine_a()
    machine_ejection = machine_load + machine_a_process_time
    machine_a_end.append(machine_ejection)
    next_start_time = machine_ejection
```

```python
for i in machine_a_end:
    print datetime.timedelta(seconds = i)
```

```python
##Next, we have to calculate what the start time
##is of Machine B. There is an if then statement
##to decide when the next process begins. Machine
##B will start a panel and take x time to finish.
##If the finish time of Machine B is less than the
##next finish time, you have to wait until that process
##finishes, else you can start immediately

machine_b_start = []
machine_b_end = []

for n, last_end_time in enumerate(machine_a_end):
    if n == 0 or last_end_time > machine_b_end[n -1]:
        next_start_time = last_end_time
    else:
        ##Executes if the machine a end time is less than the previous machine b end time
        next_start_time = machine_b_end[n-1]
    
    machine_load = next_start_time + process_load_time
    machine_b_start.append(machine_load)
    machine_b_process_time = machine_b()
    machine_ejection = machine_load + machine_b_process_time
    machine_b_end.append(machine_ejection)
    next_start_time = machine_ejection
```

```python
##Creating the DataFrame to put all this data together
df = DataFrame()
df['Panel'] = panels
df['Machine A Start'] = [datetime.timedelta(seconds = i)  for i in machine_a_start]
df['Machine A End'] = [datetime.timedelta(seconds = i)  for i in machine_a_end]
df['Machine B Start'] = [datetime.timedelta(seconds = i)  for i in machine_b_start]
df['Machine B End'] = [datetime.timedelta(seconds = i)  for i in machine_b_end]
```
